```python
import pandas as pd

lexicon_path = "C:/Users/kdzhou/source/repos/CrisisLex/lexicons"
```

```python
ch2_tb = pd.read_csv(
    f"{lexicon_path}/LonelinessT_with_scores_chi2.txt", 
    header=None, 
    sep=",", 
    names=["Word", "Chi2", "Term freq (chi2)"]
)

frq_tb = pd.read_csv(
    f"{lexicon_path}/LonelinessT_with_scores_frequency.txt", 
    header=None, 
    sep=",",
    names=["Word", "Frequency", "Term freq (freq)"]
)

pmi_tb = pd.read_csv(
    f"{lexicon_path}/LonelinessT_with_scores_pmi.txt", 
    header=None, 
    sep=",",
    names=["Word", "PMI", "Term freq (pmi)"]
)

rsv_tb = pd.read_csv(
    f"{lexicon_path}/LonelinessT_with_scores_rsv.txt", 
    header=None, 
    sep=",",
    names=["Word", "RSV", "Term freq (rsv)"]
)
```

```python
main_tb = ch2_tb

main_tb.loc[0:]
```

```python
tb = pd.merge( 
    left = ch2_tb,
    left_on = "Word",
    right = frq_tb,
    right_on = "Word",
    how = "outer"
)

tb = pd.merge( 
    left = tb,
    left_on = "Word",
    right = pmi_tb,
    right_on = "Word",
    how = "outer"
)

tb = pd.merge( 
    left = tb,
    left_on = "Word",
    right = rsv_tb,
    right_on = "Word",
    how = "outer"
)

tb.to_csv(f"{lexicon_path}/LexiconWithRSV.csv")
```
